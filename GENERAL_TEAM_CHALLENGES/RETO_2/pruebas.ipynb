```python
import base64

def decode_base64(mensaje_inv):
    
    # Decodificar la cadena Base64
    message_bytes = base64.b64decode(mensaje_inv)
    # Convertir los bytes decodificados a una cadena de texto
    message = message_bytes.decode('utf-8')
    return message
```

```python
def decode_decimal(decimal_string):
    # Dividir la cadena por espacios para obtener una lista de números decimales
    decimals = decimal_string.split()
    # Convertir cada número decimal a su correspondiente carácter ASCII
    chars = [chr(int(decimal)) for decimal in decimals]
    # Unir los caracteres en una sola cadena de texto
    message = ''.join(chars)
    return message
```

```python
def build_mapping(cipher_text, plain_text):
    # Asegurarse de que los textos tengan la misma longitud
    assert len(cipher_text) == len(plain_text)
    # Crear un diccionario vacío para el mapeo
    mapping = {}
    # Para cada par de caracteres en los textos cifrado y descifrado
    for c, p in zip(cipher_text, plain_text):
        # Si el carácter cifrado ya está en el mapeo y no se mapea al mismo carácter descifrado
        if c in mapping and mapping[c] != p:
            raise ValueError(f"Inconsistent mapping for character {c}")
        # Añadir el mapeo al diccionario
        mapping[c] = p
    return mapping

def decrypt_message(cipher_text, mapping):
    # Utilizar el mapeo para descifrar el mensaje
    return ''.join([mapping.get(c, c) for c in cipher_text])

cipher_text_example = "Yopl yo qk pytpl xy ceqyvu cueu wlkzfejue dqy zqkwflku vfyk yi wfzeuxl xy oqopfpqwflk m dqy yo elvqopl"
plain_text_example = "Esto es un texto de prueba para confirmar que funciona bien el cifrado de sustitucion y que es robusto"
cipher_text = "Iu wlkpeuoyku xyi qoquefl ykwemcp3 yo OqopfpqmyJY"

# Construir el mapeo a partir del ejemplo
mapping = build_mapping(cipher_text_example, plain_text_example)
# Descifrar el mensaje
plain_text = decrypt_message(cipher_text, mapping)
print(plain_text)
```

```python
def transposition_matrix(mensaje_cifrado, key):
    # Determinar el número de columnas y filas
    num_cols = len(key) # -> 9
    num_rows = len(mensaje_cifrado) // num_cols # -> 3

    # creamos una matriz de 2 dimensiones con 9 columnas y 3 filas
    matrix = [['' for _ in range(num_cols)] for _ in range(num_rows)]

    # Llenar la matriz con el mensaje cifrado
    index = 0
    #usamos doble bucle en filas y columnas para rellenar la matriz
    for row in range(num_rows):
        for col in range(num_cols):
            #index determina el caracter del mensaje cifrado
            if index < len(mensaje_cifrado):
                #rellenamos la matriz con el mensaje cifrado. un caracter por cada celda
                matrix[row][col] = mensaje_cifrado[index]
                index += 1#incrementamos el index para pasar al siguiente caracter

    return matrix

```

Este cifrado es un cifrado vigenere, ya que he probado por uno que por el nmumero podria ser de trasnposicion , pero no tiene sentido el mensaje( con la clave alkadi). Entonces investigamos sobre las carateriscas del cifrado vigenere, son al igual que el de transposición se usa una palabra clave para cifrar pero la diferencia es que en transpocicion solo hay un cambio de posiciones de letras y en el vigenere usa la clave para aplicar una serie de cifrados cesar al texto despalzando las letrasun numero de posicviones. Ademas el vigenere tioene patrones repetidos que transposicion tampoco , por lo que procedo a convertir las letras del mensaje y la clave a sus valores numéricos. Despues restaremos los valores de la clave de los valores del mensaje cifrado, para finalemnte convertir los valores resultantes de nuevo a letras.

```python
#funcion cambia de letras a numeros
# se le resta A para que empiece en 0 y normalizar y simplicificar las operaciones matematicas
#solo en 0 y 25
def letra_a_numero(letra):
    return ord(letra) - ord('A')

#funion que cambia de numeros a letras que se le suma A para que empiece en A
def numero_a_letra(numero):
    return chr(numero + ord('A'))


def descifrar_vigenere(mensaje_cifrado, clave):
    mensaje_descifrado = []
    longitud_clave = len(clave)
    
    for i, letra in enumerate(mensaje_cifrado):
        if letra.isalpha():  # isalpha() comprueba si el caracter es una letra
            numero_letra = letra_a_numero(letra.upper())
            #obtenemos el valor numerico de la letra de la clave
            #Se usa el modulo para que si la clave es mas corta que el mensaje, se repita
            numero_clave = letra_a_numero(clave[i % longitud_clave].upper())
            #aplicamos la regladel 26 para que si el numero es mayor que 26, se reste 26
            numero_descifrado = (numero_letra - numero_clave) % 26
            mensaje_descifrado.append(numero_a_letra(numero_descifrado))
        else:
            mensaje_descifrado.append(letra)  # Mantener caracteres no alfabéticos sin cambios
    
    return ''.join(mensaje_descifrado)

```

```python
mensaje_cifrado = "Gi vefw ui zvivltk5 in MtgecgxDBBmtieimm"
clave = "Vigenere"

# Convertir a mayúsculas y eliminar espacios para simplificar
mensaje_cifrado = mensaje_cifrado.replace(" ", "").upper()
clave = clave.upper()

mensaje_descifrado = descifrar_vigenere(mensaje_cifrado, clave)
print("Mensaje descifrado:", mensaje_descifrado)
```

```python
#!/usr/bin/env python3

import hashlib # para calcular el hash SHA-256 y otros hashes como MD5, SHA-1, SHA-512, etc
import os

# Directorio que contiene los archivos .txt
directorio = "/home/hashing3/Creds"
hash_esperado = "8a2f1de3b96eac2e0687ab9980d450b147aa3cb46ac891c724abaf757495518211ac71b16f59b92e7704ab1f6553e6f9609a977f723abca0f29b10089fe5db44"

# Recorrer los archivos con os.listdir, que lista los archivos en el directorio
for nombre_archivo in os.listdir(directorio):
    #con os.path.join se une el directorio con el nombre del archivo
    ruta_archivo = os.path.join(directorio, nombre_archivo)
    
    # Verificar si el archivo es un archivo .txt
    if nombre_archivo.endswith(".txt"):
        # Leer el contenido del archivo
        with open(ruta_archivo, 'r') as archivo:
            contenido = archivo.read()
        
        # calculo hashes de contenido del archivo y lo convierto a hexadecimal
        hash_sha512 = hashlib.sha512(contenido.encode()).hexdigest()
        print(f"Hash del archivo {nombre_archivo}: {hash_sha512}")
        
        # Comparar el hash generado con el hash esperado
        if hash_sha512 == hash_esperado:
            print(f"El archivo que contiene la clave correcta es: {nombre_archivo}")
            break
else:
    print("No se encontró el archivo con la clave correcta.")
```
