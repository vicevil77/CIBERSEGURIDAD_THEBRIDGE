```python
import matplotlib.pyplot as plt

# Datos de las vulnerabilidades
labels = ['Critical', 'High', 'Medium', 'Low', 'Info']
sizes = [11, 15, 23, 7, 75]
colors = ['#8B0000', '#FF0000', '#FFA500', '#FFFF00', '#0000FF']
explode = (0.1, 0, 0, 0, 0)  # Solo "explota" la primera porción (Critical)

# Crear el gráfico circular
plt.figure(figsize=(10, 7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)
plt.title('Análisis de Vulnerabilidades dispositivo Metasploitable')

# Añadir la leyenda
plt.legend(labels, loc="best")

# Añadir la definición de cada etiqueta
definitions = """
'Critical': Vulnerabilidad que puede ser explotada fácilmente y tiene un impacto severo en la seguridad del sistema.
'High': Vulnerabilidad que puede ser explotada con cierto esfuerzo y tiene un impacto significativo.
'Medium': Vulnerabilidad que requiere condiciones específicas para ser explotada y tiene un impacto moderado.
'Low': Vulnerabilidad que tiene un impacto bajo en la seguridad del sistema.
'Info': Información que no representa una vulnerabilidad directa pero puede ser útil para un atacante.
"""
plt.gcf().text(0.01, 0.01, definitions, fontsize=10)

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de las vulnerabilidades
labels = ['Critical', 'High', 'Medium', 'Low', 'Info']
sizes = [24, 84, 59, 8, 162]
colors = ['#8B0000', '#FF0000', '#FFA500', '#FFFF00', '#0000FF']

# Crear el gráfico de barras horizontales
fig, ax = plt.subplots(figsize=(10, 7))
bars = ax.barh(labels, sizes, color=colors)

# Añadir los valores en las barras
for bar in bars:
    width = bar.get_width()
    label_x_pos = width - 5 if width > 5 else width + 5
    ax.text(label_x_pos, bar.get_y() + bar.get_height()/2, f'{width}', va='center', ha='center', color='white', fontsize=12)

# Añadir el título
plt.title('Análisis de Vulnerabilidades')


# Crear un cuadro de texto con las definiciones
plt.gcf().text(0.02 , 0.02, fontsize=10, bbox=dict(facecolor='white', alpha=0.5))

# Mostrar el gráfico
plt.show()
```
