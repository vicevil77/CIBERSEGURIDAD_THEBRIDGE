```python
from hashlib import md5

# Certificado original obtenido con nessus o similar    
original_cert = b"Certificado original con MD5"
original_hash = md5(original_cert).hexdigest()

# Certificado falso usando tecnicas de colision, para generar un certificado falso con la misma firma digital
fake_cert = b"Certificado falso con MD5"
fake_hash = md5(fake_cert).hexdigest()

# Verificacin de colisin de hash MD5
if original_hash == fake_hash:
    print("Colisin exitosa: El certificado falso tiene la misma firma digital que el original.")
else:
    print("No se pudo generar una colisin.")
```
