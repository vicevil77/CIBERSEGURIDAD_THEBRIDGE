```python
import matplotlib.pyplot as plt
import networkx as nx

# Crear un grafo dirigido
G = nx.DiGraph()

# Añadir nodos y aristas
G.add_node("S.B.S. TELECOMUNICACIONES, S.L.")
G.add_node("Dirección General")
G.add_node("Departamento de Ventas")
G.add_node("Departamento de Marketing")
G.add_node("Departamento de Finanzas")
G.add_node("Departamento de Recursos Humanos")
G.add_node("Departamento de IT")
G.add_node("Departamento de Operaciones")

G.add_edges_from([
    ("S.B.S. TELECOMUNICACIONES, S.L.", "Dirección General"),
    ("Dirección General", "Departamento de Ventas"),
    ("Dirección General", "Departamento de Marketing"),
    ("Dirección General", "Departamento de Finanzas"),
    ("Dirección General", "Departamento de Recursos Humanos"),
    ("Dirección General", "Departamento de IT"),
    ("Dirección General", "Departamento de Operaciones")
])

# Dibujar el grafo
pos = nx.spring_layout(G)
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_weight="bold", arrows=True)
plt.title("Estructura Organizativa de S.B.S. TELECOMUNICACIONES, S.L.")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos para el gráfico
labels = ['Phishing', 'Ingeniería Social', 'Ataques Físicos', 'Suplantación de Identidad', 'Vulnerabilidades Web']
sizes = [25, 20, 15, 20, 20]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1, 0, 0, 0, 0)  # resaltar el primer segmento

# Crear el gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Vectores de Ataque Potenciales')
plt.axis('equal')  # Para que el gráfico sea un círculo
plt.show()
```
