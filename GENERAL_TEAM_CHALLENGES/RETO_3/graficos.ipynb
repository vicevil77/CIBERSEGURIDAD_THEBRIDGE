```python
import matplotlib.pyplot as plt

# Datos para el gráfico
labels = ['Phishing', 'Ingeniería Social', 'Ataques Físicos', 'Suplantación de Identidad', 'Vulnerabilidades Web']
sizes = [25, 20, 15, 20, 20]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1, 0, 0, 0, 0)  # resaltar el primer segmento

# Crear el gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Vectores de Ataque Potenciales')
plt.axis('equal')  # Para que el gráfico sea un círculo
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Crear un grafo dirigido
G = nx.DiGraph()

# Añadir nodos
nodos = [
    "Hotel (www.echaurren.com)", "Dirección General",
    "Departamento de Recepción", "Departamento de Marketing",
    "Reservas", "Atención al Cliente", "Facturación",
    "Redes Sociales", "Contenido Web", "Análisis de Datos",
    "Eventos", "Planificación"
]

for nodo in nodos:
    G.add_node(nodo)

# Añadir aristas
aristas = [
    ("Hotel (www.echaurren.com)", "Dirección General"),
    ("Dirección General", "Departamento de Recepción"),
    ("Dirección General", "Departamento de Marketing"),
    ("Departamento de Recepción", "Reservas"),
    ("Reservas", "Atención al Cliente"),
    ("Atención al Cliente", "Facturación"),
    ("Departamento de Marketing", "Redes Sociales"),
    ("Redes Sociales", "Contenido Web"),
    ("Contenido Web", "Análisis de Datos"),
    ("Departamento de Marketing", "Eventos"),
    ("Eventos", "Planificación")
]

G.add_edges_from(aristas)

# Dibujar el grafo usando spring_layout
pos = nx.spring_layout(G)
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_size=4000, node_color="lightblue", font_size=10, font_weight="bold", arrows=True, arrowstyle="-|>", arrowsize=12)
plt.title("Estructura Organizativa del Hotel (www.echaurren.com)")
plt.axis('off')  # Ocultar los ejes para un gráfico más limpio
plt.show()
```
