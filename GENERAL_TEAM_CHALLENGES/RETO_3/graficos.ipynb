```python
import matplotlib.pyplot as plt

# Datos para el grfico
labels = ['Phishing', 'Ingeniera Social', 'Ataques Fsicos', 'Suplantacin de Identidad', 'Vulnerabilidades Web']
sizes = [25, 20, 15, 20, 20]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1, 0, 0, 0, 0)  # resaltar el primer segmento

# Crear el grfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Vectores de Ataque Potenciales')
plt.axis('equal')  # Para que el grfico sea un crculo
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Crear un grafo dirigido
G = nx.DiGraph()

# Aadir nodos
nodos = [
    "Hotel (www.echaurren.com)", "Direccin General",
    "Departamento de Recepcin", "Departamento de Marketing",
    "Reservas", "Atencin al Cliente", "Facturacin",
    "Redes Sociales", "Contenido Web", "Anlisis de Datos",
    "Eventos", "Planificacin"
]

for nodo in nodos:
    G.add_node(nodo)

# Aadir aristas
aristas = [
    ("Hotel (www.echaurren.com)", "Direccin General"),
    ("Direccin General", "Departamento de Recepcin"),
    ("Direccin General", "Departamento de Marketing"),
    ("Departamento de Recepcin", "Reservas"),
    ("Reservas", "Atencin al Cliente"),
    ("Atencin al Cliente", "Facturacin"),
    ("Departamento de Marketing", "Redes Sociales"),
    ("Redes Sociales", "Contenido Web"),
    ("Contenido Web", "Anlisis de Datos"),
    ("Departamento de Marketing", "Eventos"),
    ("Eventos", "Planificacin")
]

G.add_edges_from(aristas)

# Dibujar el grafo usando spring_layout
pos = nx.spring_layout(G)
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_size=4000, node_color="lightblue", font_size=10, font_weight="bold", arrows=True, arrowstyle="-|>", arrowsize=12)
plt.title("Estructura Organizativa del Hotel (www.echaurren.com)")
plt.axis('off')  # Ocultar los ejes para un grfico ms limpio
plt.show()
```
